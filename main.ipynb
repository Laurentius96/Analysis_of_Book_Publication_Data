```python
import scipy.stats as stats 

# Separando os grupos
grupo_A = df[df["Author_Rating"] == "Famous"]["units sold"]
grupo_B = df[df["Author_Rating"] == "Intermediate"]["units sold"]

# Realizando o teste t de Student para amostras independentes
t_stat, p_val = stats.ttest_ind(grupo_A,grupo_B, equal_var=False) #equal_var=False usa a variâncias populacionais dos dois grupos comparados são iguais

# Exibir os resultados
print(f"Estatística t:{t_stat:.4f}")
print(f"Valor p: {p_val:.4f}")
```
